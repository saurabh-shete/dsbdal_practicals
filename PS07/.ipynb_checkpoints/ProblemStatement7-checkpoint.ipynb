```python
from sklearn.preprocessing import MinMaxScaler
cols = df.columns
df_copy = df.copy()
# apply normalization techniques
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled,columns=cols)
print("Scaled Dataset Using MinMaxScaler")
df_scaled.head()
```

```python
df.isna().sum()
```

```python
for i in df_scaled.columns[df_scaled.isna().any(axis=0)]:
    df_scaled[i].fillna(df_scaled[i].mean(),inplace=True)
df_scaled.isna().sum()
```

```python
X = df_scaled[['LSTAT','RM']]
Y = df_scaled['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=47)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
# model evaluation for training set
from sklearn.metrics import r2_score
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
newdf = sns.load_dataset('iris')
newdf['species'] = labelencoder.fit_transform(newdf['species'])
```

```python
newdf
```

```python
import matplotlib.pyplot as plt
from scipy import stats
fig, axs = plt.subplots(ncols=3,nrows=2,figsize=(25,15))
index = 0
axs = axs.flatten()
for col, v in newdf.items():
    sns.boxplot(y=col,ax=axs[index],data=newdf)
    index += 1
plt.show()
```

```python
sns.heatmap(newdf.corr().abs(),annot=True)
plt.show()
```

```python
newdf.columns
```

```python
X = newdf[['petal_length','petal_width','sepal_length']]
Y = newdf['species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=47)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin_mod = LinearRegression()

lin_mod.fit(X_train,Y_train)
```

```python
lin_mod.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error

y_test_predict = lin_mod.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
