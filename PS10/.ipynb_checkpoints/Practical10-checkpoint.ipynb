```python
import docx2txt
path = 'testdoc.docx'
text1 = docx2txt.process(path)
text1
```

```python
path = 'testdoc2.docx'
text2 = docx2txt.process(path)
text2
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
sent1 = sent_tokenize(text1)
sent2 = sent_tokenize(text2)
print(sent1,sent2)
```

```python
from nltk.tokenize import word_tokenize
word1 = word_tokenize(text1)
word2 = word_tokenize(text2)
print(word1,word2)
```

```python
from nltk.probability import FreqDist
freq1 = FreqDist(word1)
freq2 = FreqDist(word2)
print(freq1,freq2)
```

```python
print(f"{freq1.most_common()}            {freq2.most_common()}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,15))
freq1.plot(40) #show number of words
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
new_words1 = []
for it in word1:
    if it not in stop_words:
        new_words1.append(it)
new_words1
```

```python
new_words2 = []
for it in word2:
    if it not in stop_words:
        new_words2.append(it)
new_words2
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
stemmed_word1=[]
for it in new_words1:
    stemmed_word1.append(ps.stem(it))
print("Old Words")
print(new_words1)
print("Stemmed Word")
print(stemmed_word1)
```

```python
stemmed_word2=[]
for it in new_words2:
    stemmed_word2.append(ps.stem(it))
print("Old Words")
print(new_words2)
print("Stemmed Word")
print(stemmed_word2)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
lemmatize_word1=[]
for it in new_words1:
    lemmatize_word1.append(wnl.lemmatize(it))
print("Old World")
print(new_words1)
print("Lemmatized Word")
print(lemmatize_word1)
```
