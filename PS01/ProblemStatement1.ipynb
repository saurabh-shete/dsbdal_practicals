```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
for column in df:
    if(df[column].dtypes=="object"):
        df[column] = labelencoder.fit_transform(df[column])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
for column in df:
    if(df[column].dtypes=="object"):
        df[column] = pd.DataFrame(onehotencoder.fit_transform(df[[column]]).toarray())
df
```

```python
##Only use this to show off how much you know about DSBD otherwise don't because external will go hard on you if you use this
cols = []
for column in df:
    if(df[column].dtypes=="object"):
        cols.append(column)
df = pd.get_dummies(df,columns=cols)
df
```
