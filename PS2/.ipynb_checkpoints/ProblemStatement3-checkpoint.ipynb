```python
#checking for missing data
df.isna().sum()
```

```python
#filling the missing data
df=df.fillna(df.mean())
```

```python
df.isna().sum()
```

```python
sns.boxplot(data=df)  #displays a boxplot
```

```python
dfnew=pd.concat([df['math score'],df['reading score'],df['writing score']],axis=1) #concatenation of 
```

```python
dfnew
```

```python
#method used to remove outliers 
#IQR , Zscore etc
#We are using IQR Inter Quantile Range
Q1=dfnew.quantile(0.25) #25 percent of value
Q3=dfnew.quantile(0.75) #75 percent of value
IQR=Q3-Q1
print(IQR)
```

```python
low=Q1-1.5*IQR
high=Q3+1.5*IQR
print(low)
print(high)
```

```python
newdf=dfnew[~((dfnew<low)|(dfnew>high)).any(axis=1)]
newdf
```

```python
print(dfnew.shape)
print(newdf.shape)
```

```python
#again checking for outiers
sns.boxplot(data=newdf)
```

```python
#Normalize it using min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
newdf=scaler.fit_transform(newdf)
```

```python
newdf
```

```python
newdf=pd.DataFrame(newdf,columns=['math','reading','writing'])
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Read the dataset into a DataFrame
data = pd.read_csv('UpdatedStudentsPerformance.csv')

# Select the categorical columns to encode
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Perform one-hot encoding
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(data[categorical_cols])

# Create a DataFrame with the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names(categorical_cols))

# Concatenate the original data with the encoded data
final_data = pd.concat([data, encoded_df], axis=1)

print(final_data.head())
```
