```python
newdf1 = lassan[['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'salary']]
```

```python
newdf1.isna().sum()
```

```python
newdf1['salary'].fillna(newdf1['salary'].mean(),inplace=True)
```

```python
newdf1.isna().sum()
```

```python
newdf1.shape
```

```python
Q1 = newdf1.quantile(q=0.25)
Q3 = newdf1.quantile(q=0.75)
IQR = Q3-Q1

low = Q1- 1.5*IQR
high = Q3+1.5*IQR

print(low)
print(high)
```

```python
newdf1 = newdf1[~((newdf1<low)|(newdf1>high)).any(axis=1)]
```

```python
newdf1
```

```python
newdf1['ssc_p'] = newdf1['ssc_p'].astype('str')
```
