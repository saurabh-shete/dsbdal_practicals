```python
#creating bins randomly select some sample points
bins=[19,25,31,36,40]
labels=['19-24','25-30','31-35','37-40'] #label fot new group
df['AgeGroup']=pd.cut(df['Age'],bins=bins,labels=labels,right=False) #cut is use to select values and create bin
```

```python
df.groupby('AgeGroup')['Salary'].mean()  #mean salary by age group
```
